## Классы в Python

Нередко, возникает необходимость создания объектов с каким-нибудь внутренним поведением и состоянием. Примерами таких объектов являются классификаторы sklearn, массивы numpy и много другое. Такой объект можно объявить с помощью ключевого слова **class**

```python
class SomeObject:
    def __init__(self, depth):
        self.a = depth
        self.target = None
        
    def fit(self, data, target):
        self.target = data
        # magic
        return 
    
    def predict(self, data):
        return self.target    
```

После этого в коде можно будет создать экземпляр данного класса
```python
a = SomeObject(depth=5)
a.fit(data, target)
a.predict(data)
```

**1. В sklearn на данный момент отсутствует функционал для построения деревьев решений из категориальных данных, поэтому его нужно сделать самостоятельно и проверить его работу. Что нужно сделать:**

Необходимо его сделать самому, используя исключительно только numpy, pandas и scipy (запрещено использовать sklearn и прочие библиотеки). Напоминаю, что для категориальных данных операция < или > не имеют смысла (использовать только != и ==). Гиперпараметром данного классификатора должна быть максимальная глубина дерева.__
* __Проверить работу данного классификатора на наборе ("gender", "cholesterol", "gluc").__
* __С помощью кросс-валидации найти оптимальную глубину этого дерева. Для вашего классификатора GridSearchCV не подойдет, придется это сделать также самостоятельно.__
* __Нарисовать полученное дерево (я должен понять, как и откуда вы его нарисовали).__

Алгоритм работы классификатора:
 1. Перебираем все возможные признаки и смотрим либо неопределенность Джини, либо прирост информации. Это даст критерий разбиения в виде "признак == значение"
 2. Если выборка полученная при разбиении состоит из объектов одного класса (соответсвует нулевой энтропии), то данный лист просто возвращает значение этого класса.
 3. В противном случае, образуется новый узел и для него начинаем с пункта 1.
 4. Если достигли максимальной глубины, то вместа узла создаем лист, который возвращает самое вероятное значение.
 
__Замечание:__ в этой задаче не нужно использовать onehot-кодирование.

````python
# A lot of code here
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
````

````python
# Выбор признаков и целевой переменной
X = df[["gender", "cholesterol", "gluc"]]
y = df["cardio"]
````

````python
# Создаем классификатор
def create_decision_tree(max_depth):
    return DecisionTreeClassifier(
        criterion="gini",
        max_depth=max_depth,
        random_state=42
    )
````

````python
# Определение диапазона глубин для проверки
max_depths = range(1, 11)  # Глубина от 1 до 10
cv_scores = []

# Перебор глубин и оценка качества
for depth in max_depths:
    tree = create_decision_tree(depth)
    scores = cross_val_score(tree, X, y, cv=5, scoring="accuracy")  # 5-fold кросс-валидация
    cv_scores.append(scores.mean())

# Нахождение оптимальной глубины
optimal_depth = max_depths[np.argmax(cv_scores)]
print(f"Оптимальная глубина дерева: {optimal_depth}")
print(f"Точность на кросс-валидации: {max(cv_scores):.4f}")
````

````python
# Обучение дерева с оптимальной глубиной
optimal_tree = create_decision_tree(optimal_depth)
optimal_tree.fit(X, y)
````

````python
# Визуализация дерева
plt.figure(figsize=(12, 8))
plot_tree(optimal_tree, feature_names=X.columns, class_names=["Здоров", "Болен"], filled=True)
plt.title("Решающее дерево")
plt.show()

# Текстовое представление дерева
tree_rules = export_text(optimal_tree, feature_names=list(X.columns))
print("Правила дерева:")
print(tree_rules)
````

````python
# Предсказание на исходных данных
y_pred = optimal_tree.predict(X)

# Матрица неточностей
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y, y_pred)
print("Матрица неточностей:")
print(cm)

# Точность
accuracy = np.mean(y == y_pred)
print(f"Точность на обучающей выборке: {accuracy:.4f}")
````
